# Modeling

The purpose of this notebook is to develop and evaluate machine learning models capable of predicting next year's California Dashboard color classification for Aspire Public Schools.

The preprocessing notebook prepared the dataset by:
- creating the future prediction target,
- engineering features,
- encoding categorical variables,
- standardizing numeric variables,
- and splitting the data into training and testing datasets.

This notebook will:
- train multiple classification models,
- evaluate model performance,
- compare results across models,
- and identify the final model for the project.

```python
# Import libraries

import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix
)
```

```python
# Load preprocessed datasets

X_train = pd.read_csv("X_train_scaled.csv")
X_test = pd.read_csv("X_test_scaled.csv")

y_train = pd.read_csv("y_train.csv").squeeze()
y_test = pd.read_csv("y_test.csv").squeeze()

# Check dataset shapes

print("X_train:", X_train.shape)
print("X_test:", X_test.shape)

print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

## Baseline Model: Logistic Regression

The first model used in this project is Logistic Regression. Logistic Regression is commonly used as a baseline classification model because it is relatively simple, interpretable, and computationally efficient.

This model will provide a benchmark for comparing the performance of more complex ensemble models later in the project.

```python
# Create Logistic Regression model

log_reg = LogisticRegression(
    max_iter=5000,
    multi_class='multinomial'
)

# Fit model on training data

log_reg.fit(X_train, y_train)

# Generate predictions

y_pred_log = log_reg.predict(X_test)
```

```python
# Evaluate Logistic Regression model

print("Logistic Regression Accuracy:")
print(accuracy_score(y_test, y_pred_log))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_log))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_log))
```

## Logistic Regression Results

The Logistic Regression model achieved an accuracy of approximately 36.6% on the testing dataset.

The model performed best when predicting Dashboard color class 3, but struggled to accurately classify several of the smaller classes, particularly classes 1, 2, and 4.

This outcome suggests that the relationship between the predictor variables and Dashboard color classifications may be nonlinear and more complex than Logistic Regression can effectively capture.

Because Logistic Regression is a relatively simple linear model, it serves primarily as a useful baseline for comparison against more advanced ensemble methods.

## Model 2: Random Forest Classifier

The next model used is a Random Forest Classifier.

Random Forest is an ensemble learning method that combines many decision trees to improve predictive performance and reduce overfitting. Unlike Logistic Regression, Random Forest can capture nonlinear relationships and interactions between variables.

This model may perform better on the Aspire Dashboard dataset because educational performance outcomes are likely influenced by complex combinations of historical indicators and trends.

```python
# Create Random Forest model

rf_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=5,
    random_state=42
)

# Fit model

rf_model.fit(X_train, y_train)

# Generate predictions

y_pred_rf = rf_model.predict(X_test)
```

```python
# Evaluate Random Forest model

print("Random Forest Accuracy:")
print(accuracy_score(y_test, y_pred_rf))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))
```

## Random Forest Results

The Random Forest model achieved an accuracy of approximately 33.8%, which was slightly lower than the Logistic Regression baseline model.

Like Logistic Regression, the Random Forest model performed best when predicting Dashboard color class 3, while struggling with several of the smaller classes.

Although Random Forest is capable of modeling nonlinear relationships, the relatively small dataset size may limit the model’s ability to generalize effectively. Ensemble tree models often perform best with larger training datasets.

Despite the lower overall accuracy, Random Forest remains valuable because it can capture more complex feature interactions and may provide useful feature importance insights later in the analysis.

## Model 3: Gradient Boosting Classifier

The final model evaluated is a Gradient Boosting Classifier.

Gradient Boosting builds trees sequentially, with each new tree attempting to correct the errors made by previous trees. This often produces stronger predictive performance than individual decision trees or standard Random Forest models.

Because Gradient Boosting can model complex nonlinear relationships while reducing bias, it is frequently one of the strongest-performing algorithms for structured tabular datasets.

```python
# Create Gradient Boosting model

gb_model = GradientBoostingClassifier(
    n_estimators=100,
    learning_rate=0.05,
    max_depth=3,
    random_state=42
)

# Fit model

gb_model.fit(X_train, y_train)

# Generate predictions

y_pred_gb = gb_model.predict(X_test)
```

```python
# Evaluate Gradient Boosting model

print("Gradient Boosting Accuracy:")
print(accuracy_score(y_test, y_pred_gb))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_gb))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_gb))
```

## Gradient Boosting Results

The Gradient Boosting model achieved an accuracy of approximately 36.6%, matching the Logistic Regression model and outperforming the Random Forest model slightly.

Gradient Boosting showed improved balance across several Dashboard color classes compared to the previous models, particularly for classes 2 and 4. The model again performed best when predicting Dashboard color class 3, which was also the largest class in the dataset.

Overall, the results suggest that the dataset contains meaningful predictive signal, but the relatively small sample size and multiclass structure make accurate prediction challenging.

Because Gradient Boosting can capture nonlinear relationships and interactions between variables while maintaining competitive accuracy, it appears to be one of the strongest candidate models for this project.

# Compare Model Performance

The three classification models produced relatively similar performance results overall.

| Gradient Boosting | 36.6% |

Logistic Regression and Gradient Boosting achieved the highest accuracy scores, while Random Forest performed slightly worse on the testing dataset.

Although Logistic Regression matched Gradient Boosting in overall accuracy, Gradient Boosting is likely the stronger final model because:

- it handles feature interactions more effectively,

The relatively modest performance across all models is likely influenced by:
- the small dataset size,

# Final Model Selection

The final model selected for this project is the Gradient Boosting Classifier.

Gradient Boosting was selected because it:
- achieved competitive accuracy,
- handled multiclass prediction reasonably well,
- captured nonlinear relationships between educational performance indicators,
- and demonstrated more balanced classification behavior than the other evaluated models.

While overall predictive performance remains moderate, the model still provides valuable insight into the relationship between historical Dashboard indicators and future school accountability outcomes.
